```python
import pandas as pd
hw_data = pd.read_csv('mapped_plant_quads_co_allrecords_cover.csv')
```

```python
"""
check all the data sets then I could see which ones to merge to get a completed one
"""
hw_data2= pd.read_csv('mapped_plant_quads_co_allrecords_density.csv')
hw_data2.head()
```

```python
hw_data3= pd.read_csv('mapped_plant_quads_co_daily_climate.csv')
hw_data3.head()
```

```python
hw_data4= pd.read_csv('mapped_plant_quads_co_quad_info.csv')
hw_data4.head()
```

```python
hw_data5= pd.read_csv('mapped_plant_quads_co_quad_inventory.csv')
hw_data5.head()
```

```python
hw_data6= pd.read_csv('mapped_plant_quads_co_quad_stocking_rate.csv')
hw_data6.head()
```

```python
hw_data7= pd.read_csv('mapped_plant_quads_co_species_list.csv')
hw_data7.head()
```

```python
hw_data8= pd.read_csv('mapped_plant_quads_co_species_name_changes.csv')
hw_data8.head()
```

```python
"""
merge these two to get a dataset with quad, year species area and growthform
"""
merged1= pd.merge(left = hw_data, right = hw_data7, left_on = 'species',
                       right_on = 'species',how = 'outer')
merged1.head()#outer merge to include all data
```

```python
merged1.species.unique()#check values for each variable
merged1.quad.unique()
```

```python
"""
merge these two to get the data with quad, year, species, area, growthform and precipitation
"""

merged2= pd.merge(left = merged1, right = hw_data3, left_on = 'year',
                       right_on = 'year',how = 'outer')
merged2.head()
```

```python
"""
Q1 A plot of the area covered by each species over all quadrats (y) for each year (x).
beautiful graph with Y-axis counts of each species and x-axis year
there are two species have most of counts, orange and green ones. But some colors looks so similar so it is hard to tell
"""
quad_plot = merged1.groupby(['species','year'], 
                         as_index = False).area.sum()
```

```python
quad_plot_wide = quad_plot.pivot(index = 'year', columns = 'species', 
                             values = 'area')
```

```python
quad_plot_wide.head()
```

```python
plot1= quad_plot_wide.plot(kind = 'bar', stacked = True, 
                  title = "quad counts")
plot1.set_xlabel("year")
plot1.set_ylabel("Counts")
```

```python
"""
Q2 A plot of the area covered by each growth-form over all quadrats (y) for each year (x).
beautiful graph with Y-axis counts of each growthform and x-axis year
not plant and grass have most of counts
"""

quad_plot2= merged1.groupby(['growthform','year'], #group by growthform and year
                         as_index = False).area.sum()
quad_plot_wide2= quad_plot2.pivot(index = 'year', columns = 'growthform', 
                             values = 'area')
plot2= quad_plot_wide2.plot(kind = 'bar', stacked = True, 
                  title = "quad counts")
plot2.set_xlabel("year")
plot2.set_ylabel("Counts")  
```

```python
# Q3 A plot of the area covered by each growth-form over all quadrats (y) as a function of the mean precipitation for the year (x).
```

```python
"""
# Q3 A plot of the area covered by each growth-form over all quadrats (y) as a function of the mean precipitation for the year (x).
beautiful graph considers mean precipitation
"""

quad_plot3= merged2.groupby(['year','growthform'], 
                         as_index = False).precip.mean()
quad_plot_wide3= quad_plot3.pivot(index = 'year', columns = 'growthform', 
                             values = 'precip')
plot3= quad_plot_wide3.plot(kind = 'bar', stacked = True, 
                  title = "quad counts")
plot3.set_xlabel("year")
plot3.set_ylabel("Counts")
```

```python
"""
#Q4 A plot of the area covered by each longevity over all quadrats (y) for each year (x).

"""

quad_plot2= merged1.groupby(['longevity','year'], 
                         as_index = False).area.sum()
quad_plot_wide2= quad_plot2.pivot(index = 'year', columns = 'longevity', 
                             values = 'area')
plot2= quad_plot_wide2.plot(kind = 'bar', stacked = True, 
                  title = "quad counts")
plot2.set_xlabel("year")
plot2.set_ylabel("Counts")
```
